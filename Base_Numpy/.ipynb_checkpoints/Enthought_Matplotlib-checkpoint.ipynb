```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plotting a line
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x));
#; this is to remove [<matplotlib.lines.Line2D at 0x2923fc1def0>] that shows up at the top
```

```python
#less stepsin the linspace function (from 100 to 10) - less smooth
x = np.linspace(0, 10, 10)
plt.plot(x, np.sin(x));
```

```python
#scatter plot points 
x = np.random.normal(size = 500)
y = np.random.normal(size = 500)
plt.scatter(x,y);
```

```python
plt.hist(x, bins = 50);
```

```python
#Showing images using imshow 
# - note that origin is at the top-left by default 

x = np.linspace(1, 12, 100)
y = x[:, np.newaxis]

im = y * np.sin(x) * np.cos(y)
print(im.shape)

plt.imshow(im);
```

```python
#Contour plots 
# - note that origin here is at the bottom-left by default 
plt.contour(im);
```

```python
plt.contourf(im);
```
