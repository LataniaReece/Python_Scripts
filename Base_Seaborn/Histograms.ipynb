```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import style 

style.use('ggplot')

%matplotlib inline
```

```python
num = np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num, color = 'blue') 
```

```python
#giving the variable a name 
label_dist = pd.Series(num, name = 'variable x')
```

```python
#now the name will appear on the histogram
sns.distplot(label_dist)
```

```python
#make it a vertical plot 
sns.distplot(label_dist, vertical = True)
```

```python
#remove the histogram and just keep the line - similar to KDE plot (kernel density function)
sns.distplot(label_dist, hist = False)
```

```python
#adding a rug to the plot 
sns.distplot(label_dist, rug = True, hist = False)
```
