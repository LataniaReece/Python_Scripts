```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import style 
style.use('seaborn')

%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
x = sns.jointplot(tips.total_bill, tips.tip)
```

```python
from scipy.stats import pearsonr
pearsonr(tips.total_bill, tips.tip)
```

```python
#stat_func depracated in jointplot so you would need to add it like this
x = sns.jointplot(tips.total_bill, tips.tip)
plt.annotate(pearsonr(tips.total_bill, tips.tip), xy = (10,10))
plt.show()
```

```python
sns.jointplot(iris.sepal_length, iris.sepal_width)
```

```python
#adding regression line to plot
sns.jointplot(tips.total_bill, tips.tip, color = 'purple', kind = 'reg')
```

```python
#hexagon scatterplot
sns.jointplot(tips.total_bill, tips.tip, color = 'purple', kind = 'hex')
```

```python
#kde scatterplot
sns.jointplot(tips.total_bill, tips.tip, color = 'purple', kind = 'kde')
```

```python
#iris dataset kde
sns.jointplot(iris.petal_length, iris.petal_width,kind = 'kde')
```
