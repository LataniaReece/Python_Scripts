```python
import numpy as np 
import pandas as pd 

import seaborn as sns 
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('ggplot')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.boxplot(tips['size'])
#didn't work: sns.boxplot(tips.size) - size is an actual function in python
#When I changed it to a different name it worked - see below
```

```python
test = tips.rename(columns = {'size':'sizes'})
sns.boxplot(test.sizes)
```

```python
sns.boxplot(tips.total_bill)
```

```python
# x and y boxplot 
sns.boxplot(tips.sex, tips.total_bill)
```

```python
#you can also change the order 
sns.boxplot(tips.sex, tips.total_bill, order = ['Female', 'Male'], palette = "coolwarm")
```

```python
#using two categories in a boxplot 
sns.boxplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'husl')
```

```python
#using iris data set to show you can just use df as input to boxplot function
iris = sns.load_dataset('iris')
type(iris)
```

```python
#sns.boxplot(iris) did not work for some reason 
sns.boxplot(data=iris, palette = 'coolwarm')
```

```python
#boxplot orientation 
sns.boxplot(data=iris, palette = 'husl', orient = 'horizantal')
```

```python
#adding data points to the boxplot - need to do both of these together 
sns.boxplot(tips.day, tips.total_bill, palette = 'husl')
sns.swarmplot(tips.day, tips.total_bill)
```

```python
#change the color of the dots
sns.boxplot(tips.day, tips.total_bill, palette = 'husl')
sns.swarmplot(tips.day, tips.total_bill, color = 'black')
```

```python
#you can use a number to change the color 
sns.boxplot(tips.day, tips.total_bill, palette = 'husl')
sns.swarmplot(tips.day, tips.total_bill, color = "0.6") #darker colors are closer to 0, brighter colors are higher 
```
