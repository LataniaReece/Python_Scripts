```python
import numpy as np 
import pandas as pd 

import seaborn as sns 
import matplotlib.pyplot as plt 
from matplotlib import style
style.use('seaborn')

%matplotlib inline
```

```python
#facet allows you to see the graphs for different types of observations in your dataset 
#ie the first one whows total bill for each different combo of smoker and time 
```

```python
tips = sns.load_dataset('tips')
```

```python
x = sns.FacetGrid(tips, row = 'smoker', col = 'time')
x = x.map(plt.hist, 'total_bill', edgecolor = 'black')
```

```python
#changing the color and the bins
x = sns.FacetGrid(tips, row = 'smoker', col = 'time')
x = x.map(plt.hist, 'total_bill', edgecolor = 'black', color = 'brown', bins = 15)
```

```python
#scatterplots
x = sns.FacetGrid(tips, col = 'time', row = 'smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
#adding category color to the scatterplot
x = sns.FacetGrid(tips, col = 'time', hue = 'smoker')
x = x.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
#changing pallete
x = sns.FacetGrid(tips, col = 'time', hue = 'smoker', palette = "Blues_r")
x = x.map(plt.scatter, 'total_bill', 'tip', edgecolor = "black")
```

```python
#regressionplot
x = sns.FacetGrid(tips, col = 'time', row = 'smoker')
x = x.map(sns.regplot, 'total_bill', 'tip')
```

```python
#Boxplot
x = sns.FacetGrid(tips, col ='day')
x = x.map(sns.boxplot, 'total_bill', 'time')
```

```python
#switching the order of time and total_bill
x = sns.FacetGrid(tips, col ='day')
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
#changing the size of the plots 
x = sns.FacetGrid(tips, col ='day', height =7, aspect = 0.4)
x = x.map(sns.boxplot, 'time', 'total_bill')
```

```python
#you can also change the order 
x = sns.FacetGrid(tips, col ='day', height =7, aspect = 0.4, col_order = ['Sat', 'Thur', 'Sun', 'Fri'])
x = x.map(sns.boxplot, 'time', 'total_bill')
```
