```python
import numpy as np 
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plot 
from matplotlib import style 
style.use('seaborn')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.violinplot(tips.tip)
```

```python
sns.violinplot(tips.day, tips.total_bill)
```

```python
#adding sex category
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'husl')
```

```python
#splitting the violins by category
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'husl', split = True)
```

```python
#just like in others you can change the order 
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'husl', split = True, 
               order = ['Sat', 'Thur', 'Fri', 'Sun'])
```

```python
#Changing the way the inner violin looks, this time it's lines representaing the quartiles 
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'coolwarm',
              inner = 'quartile')
```

```python
#adding split = True again 
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'coolwarm',
              inner = 'quartile', split = True)
```

```python
#width of the violin means there were less observations in that category
sns.violinplot(tips.day, tips.total_bill, hue = tips.sex, palette = 'RdBu',
              inner = 'quartile', scale = 'count')
```

```python
#another example - again smaller means less observations in that example
sns.violinplot(tips.day, tips.total_bill, hue = tips.smoker, palette = 'RdBu',
              inner = 'quartile', scale = 'count')
```

```python
#instead of using the scale, you can also use the lines, more lines means more observations
#inner = 'stick'
sns.violinplot(tips.day, tips.total_bill, hue = tips.smoker, palette = 'RdBu',
              inner = 'stick')
```

```python
sns.violinplot(tips.day, tips.total_bill, hue = tips.smoker, palette = 'RdBu',
              inner = 'stick', split = True)
```

```python
#you can also compare across the x axis scale wise 
#scale_hue = False
sns.violinplot(tips.day, tips.total_bill, hue = tips.smoker, palette = 'husl',
              inner = 'stick', scale = 'count', scale_hue = False)
```

```python
#adding split 
#scale_hue = False
sns.violinplot(tips.day, tips.total_bill, hue = tips.smoker, palette = 'husl',
              inner = 'stick', scale = 'count', scale_hue = False, split = True)
```
