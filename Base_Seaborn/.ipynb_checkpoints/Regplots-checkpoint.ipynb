```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('seaborn')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot(tips.total_bill, tips.tip)
```

```python
#simulating data with better regression - used in the video
#https://youtu.be/g2M1UgjtdNI?list=PL998lXKj66MpNd0_XkEXwzTGPxY2jYM2d
mean = [2,5]
cov = [[1.1, 0.4], [2.2, 3]]

x_value, y_value = np.random.multivariate_normal(mean, cov, 100).T

sns.regplot(x_value, y_value)
```

```python
#just adding names to the plot
series_x_value = pd.Series(x_value, name = "Var-X")
series_y_value = pd.Series(y_value, name = "Var-Y")

sns.regplot(series_x_value, series_y_value)
```

```python
#changing the marker and color 
sns.regplot(series_x_value, series_y_value, marker = 'D', color = 'gray' )
```

```python
#changing line and dots , using tips data 
sns.regplot(tips.total_bill, tips.tip,
           scatter_kws = {'marker': 'D', 'color':'blue'},
           line_kws = {'color':'red', 'linewidth':3.1})
```

```python
sns.regplot(x_value, y_value) #reminder
```

```python
#changing the confidence interval of the regression line 
sns.regplot(x_value, y_value, ci = 75)
```

```python
sns.regplot(tips['size'], tips.total_bill)
```

```python
#add jitter to plot above
sns.regplot(tips['size'], tips.total_bill, x_jitter = 0.1)
```

```python
#to add more jitter, increase the number 
sns.regplot(tips['size'], tips.total_bill, x_jitter = 0.3)
```

```python
#less jitter, decrease number
sns.regplot(tips['size'], tips.total_bill, x_jitter = 0.08)
```

```python
#changing color of scatter points
sns.regplot(tips['size'], tips.total_bill, x_jitter = 0.08, scatter_kws ={'color': 'red'})
```

```python
#adding the mean to the chart 
sns.regplot(tips['size'], tips.total_bill, x_estimator = np.mean)
```
