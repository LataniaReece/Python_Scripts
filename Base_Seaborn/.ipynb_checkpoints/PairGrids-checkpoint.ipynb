```python
import numpy as np 
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import style
style.use('fivethirtyeight')

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
x = sns.PairGrid(iris)
x = x.map(plt.scatter)
```

```python
#adds histograms on the diagonals 
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
#adding species color to the plots
x = sns.PairGrid(iris, hue = 'species', palette = 'husl')
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
#changing histogram type 
x = sns.PairGrid(iris, hue = 'species', palette = 'husl')
x = x.map_diag(plt.hist, histtype = "step", linewidth = 4)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
#choosing which variables you want in the pair plot
x = sns.PairGrid(iris, vars = ['petal_length', 'petal_width'])
x = x.map(plt.scatter)
```

```python
#adding species
x = sns.PairGrid(iris, hue = 'species', vars = ['petal_length', 'petal_width'])
x = x.map(plt.scatter)
x = x.add_legend()
```

```python
#changing to histogram on diagonals and adding edgecolor 
x = sns.PairGrid(iris, hue = 'species', vars = ['petal_length', 'petal_width'])
x = x.map_diag(plt.hist, edgecolor = 'black')
x = x.map_offdiag(plt.scatter, edgecolor = 'white')
x = x.add_legend()
```

```python
#you can choose the x and y axis 
x = sns.PairGrid(iris, x_vars = ['petal_length', 'petal_width'],
                y_vars=['sepal_length', 'sepal_width'])
x = x.map(plt.scatter)
```

```python
#you can change the plots that show up in upper and lower portions of pairgrid 
x = sns.PairGrid(iris)
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
```

```python
#adding species
x = sns.PairGrid(iris, hue = 'species', palette = 'husl')
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```

```python
#you can also change the marker type for each species
x = sns.PairGrid(iris, hue = 'species', hue_kws = {'marker': ['D', '*', '+']})
x = x.map(plt.scatter, s = 30, edgecolor = 'black')
x = x.add_legend()
```
