```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style 
style.use('seaborn')

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
#automatically does scatterplots and histograms 
sns.pairplot(iris)
```

```python
#tips data
x = sns.pairplot(tips)
```

```python
#hue 
sns.pairplot(iris, hue = 'species')
```

```python
#adding bars to diagonals, changing palette
sns.pairplot(iris, hue = 'species', diag_kind = 'hist', palette = 'husl')
```

```python
#adding markers and edgecolor to scatterplots 
sns.pairplot(iris, hue = 'species', palette = 'husl', markers = ['>', 's', '+'], 
             plot_kws = {'edgecolor': 'black'})
```

```python
#choosing variables, adding edgecolor to histogram
sns.pairplot(iris, height = 4, 
             vars = ['sepal_length', 'sepal_width'], diag_kws = {'edgecolor': 'black'})
```

```python
#choosing both x and y variables explicitly 
sns.pairplot(iris, x_vars = ['petal_length', 'petal_width'],
             y_vars = ['sepal_length', 'sepal_width'])
```

```python
#adding hue 
sns.pairplot(iris, x_vars = ['petal_length', 'petal_width'],
             y_vars = ['sepal_length', 'sepal_width'],
            palette = 'husl', hue = 'species', plot_kws = {'edgecolor': 'black'})
```

```python
#adding regression line to plots 
sns.pairplot(tips, kind = 'reg', diag_kws = {'edgecolor': 'black'})
```
