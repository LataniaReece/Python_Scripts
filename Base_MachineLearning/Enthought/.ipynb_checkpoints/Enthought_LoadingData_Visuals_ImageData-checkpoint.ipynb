## Iris Data

```python
from sklearn.datasets import load_iris
iris = load_iris()
# iris = bunch object which is a dictionary like object containing details about the data
#type(iris)
```

```python
iris.keys()
```

```python
n_samples, n_features = iris.data.shape
print('Number of samples:', n_samples)
print('Number of features:', n_features)
# the sepal length, sepal width, petal length and petal width of the first row
print(iris.data[0])
```

```python
print(iris.target)
print(iris.target_names)
print(iris.target.dtype)
```

#### NOTE: for scikit learn, if you give target as integers it will be interpreted as classification problem. If you give target as float, it will be interpreted as regression problem.

### Value counts for numpy 

```python
import numpy as np
np.bincount(iris.target)
```

### Visualiziations 
We can visualize one or two of the dimensions at a time using a simple histogram or scatterplot

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# In this case, we took the third columnd and plot the histogram for each of the classes 

x_index = 3

for label in range(len(iris.target_names)):
    plt.hist(iris.data[iris.target == label, x_index],
             label = iris.target_names[label],
             alpha = 0.5) # alpha = opacity of the colors 
    
plt.xlabel(iris.feature_names[x_index])
plt.legend(loc='upper right')
plt.show()
```

```python
x_index = 3
y_index = 0

for label in range(len(iris.target_names)):
    plt.scatter(iris.data[iris.target == label, x_index],
               iris.data[iris.target == label, y_index],
               label = iris.target_names[label])
    
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.legend(loc = 'upper left')
plt.show()
```

### Feature Plot

To see which features better separate the classes use a featureplot to see those relationships

```python
import pandas as pd
import mglearn

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
pd.plotting.scatter_matrix(iris_df, c = iris.target, figsize = (15,15), marker = 'o', 
                           hist_kwds = {'bins': 20, "edgecolor": 'black'}, s = 60,
                          alpha = .8, cmap = mglearn.cm3, edgecolor = 'black');
```

## Handwritten-Digits DataSet

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.keys()
```

```python
digits.data.shape
```

```python
print(digits.target_names)
print(np.bincount(digits.target))
```

```python
#Difference between the data and the images 
print(digits.data.shape) #vectorized version
print(digits.images.shape)
```

### Visualizing the image data

```python
#Set up the figure 
fig = plt.figure(figsize = (6,6)) # figure size in inches 
fig.subplots_adjust(left = 0 , right = 1, bottom = 0, top = 1, hspace = 0.05, wspace = 0.01 )

#plot the digits: each image is 8 x 8 pixels
for i in range(64): #taking the first 64 images
    ax = fig.add_subplot(8, 8, i + 1, xticks = [], yticks = [])
    ax.imshow(digits.images[i], cmap = plt.cm.binary, interpolation = 'nearest')
    
    #label the image with the target value 
    ax.text(0, 7, str(digits. target[i]))
    
    #plt.cm.binary is a good colormap for this data
```
