# Neural Networks (120)

Tuning Parameters: 
	• Number of nodes in hidden layer ( hidden_layer_sizes)
		○ Depends on the size of the data (just know that 10,000 is not far fetched if you have very complex data and 10 could be for small data)
		○ Reducing nodes in layer = reduces complexity of model 
			§ Default is 100 
		○ Often similar to the number of input features, rarely higher than in the low to mid-thousands
	• Hidden Layers 
		○ Start with one or two and then build from there 
	• Nonlinearity Function 
		○ Relu or tanh
	• Alpha 
		○ L2 reguralization (higher alpha - more reg - less complexity - less overfitting)
		






Learning algorithms in nnet ( how they learn) 
	• Adam (recommended when you start with nnet) 
		○ Works well is in most cases, sensitive to data scaling - make sure to center and scale 
	• Lbfgs  (recommended when you start with nnet) 
		○ Quite robust but takes a long time on larger datasets 
	• Sgd 
Comes with many additional parameters 

## Two moons dataset and NNET Tuning

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,  
                                                    random_state=42)
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3) 
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Default of hidden unit nodes = 100, we can reduce to 10 ( reduces model complexity)

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[10])
mlp.fit(X_train, y_train) 
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
plt.xlabel("Feature 0") 
plt.ylabel("Feature 1")
```

As you can see this doesn't affect the decision boundary much for this dataset.

### Smoothness of decision boundary 

To improve the smoothness of the decision boundary, lets add another hidden unit

#### Adding more hidden layers

```python
# using two hidden layers, with 10 units each 
mlp = MLPClassifier(solver='lbfgs', random_state=0, 
                    hidden_layer_sizes=[10, 10])
mlp.fit(X_train, y_train) 
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3)
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train) 
plt.xlabel("Feature 0") 
plt.ylabel("Feature 1")
```

Default activation function is relu, let's use the tanh activation function

#### Changing activation function to tanh

```python
# using two hidden layers, with 10 units each, now with tanh nonlinearity 
mlp = MLPClassifier(solver='lbfgs', activation='tanh',   
                    random_state=0, hidden_layer_sizes=[10, 10]) 
mlp.fit(X_train, y_train)
mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3) 
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train) 
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

### Alpha reguralization 

Remember increasing alpha means more reguralization means less model complexity, less overfitting

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for axx, n_hidden_nodes in zip(axes, [10, 100]):  
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0,
                            hidden_layer_sizes=[n_hidden_nodes, n_hidden_nodes],  
                            alpha=alpha)  
        mlp.fit(X_train, y_train)    
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)   
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)   
        ax.set_title("n_hidden=[{}, {}]\nalpha={:.4f}".format(             
            n_hidden_nodes, n_hidden_nodes, alpha))
```

## The effects of Neural Networks random initialization: 
NNET randomly selects weights to begin with, so you may get differnt results (especially if the datasize is small)

Below are plots of several models, all with the same setting parameters, but different random initializations

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 8)) 
for i, ax in enumerate(axes.ravel()):  
    mlp = MLPClassifier(solver='lbfgs', random_state=i,  
                        hidden_layer_sizes=[100, 100])   
    mlp.fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)  
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
```

## Breast Cancer Data set and NNET

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

print("Cancer data per-feature maxima:\n", cancer.data.max(axis=0))
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
    cancer.data, cancer.target, random_state=0)

mlp = MLPClassifier(random_state=42) 
mlp.fit(X_train, y_train)

print("Accuracy on training set: {:.2f}".format(mlp.score(X_train, y_train))) 
print("Accuracy on test set: {:.2f}".format(mlp.score(X_test, y_test)))
```

The results are good, but not as good as SVC which was completed in SVM notebook. This may be due to scaling issues, let's scale the data. Remember there is an easier way to do this but they wanted to show how it looks by hand and I just copied and paste

```python
# compute the mean value per feature on the training set 
mean_on_train = X_train.mean(axis=0)
# compute the standard deviation of each feature on the training set 
std_on_train = X_train.std(axis=0)

# subtract the mean, and scale by inverse standard deviation # afterward, mean=0 and std=1 
X_train_scaled = (X_train - mean_on_train) / std_on_train 
# use THE SAME transformation (using training mean and std) on the test set
X_test_scaled = (X_test - mean_on_train) / std_on_train

mlp = MLPClassifier(random_state=0) 
mlp.fit(X_train_scaled, y_train)

print("Accuracy on training set: {:.3f}".format(  
    mlp.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```

increase number of iterations

```python
mlp = MLPClassifier(max_iter=1000, random_state=0) 
mlp.fit(X_train_scaled, y_train)

print("Accuracy on training set: {:.3f}".format(  
    mlp.score(X_train_scaled, y_train))) 
print("Accuracy on test set: {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```

Good. now increase alpha to 1 to increase generalization

```python
mlp = MLPClassifier(max_iter=1000, alpha=1, random_state=0)
mlp.fit(X_train_scaled, y_train)

print("Accuracy on training set: {:.3f}".format(
    mlp.score(X_train_scaled, y_train))) 
print("Accuracy on test set: {:.3f}".format(mlp.score(X_test_scaled, y_test)))
```

### Heat Map of Coefficients in breast cancer data set 

The rows in this plot correspond to the 30 input features, while the columns correspond to the 100 hidden units. Light colors represent large positive values, while dark colors represent negative values:

```python
plt.figure(figsize=(20, 5)) 
plt.imshow(mlp.coefs_[0], interpolation='none', cmap='viridis')
plt.yticks(range(30), cancer.feature_names) 
plt.xlabel("Columns in weight matrix") 
plt.ylabel("Input feature") 
plt.colorbar();
```
